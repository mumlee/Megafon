# 1. Постановка задачи

Как и любой бизнес, компания «Мегафон» хочет повысить удовлетворённость своих клиентов качеством услуг. Это важная задача для удержания пользователей — как давних, так и недавно привлечённых. Ведь затраты на маркетинг и продвижение не оправдаются, если клиент уйдёт из-за низкого качества связи. Однако в реальном мире ресурсы всегда ограничены, и в единицу времени технический отдел может решить конечное количество задач. 

Чтобы делать это наиболее эффективно, важно определить, какие технические показатели качества связи сильнее всего влияют на удовлетворённость клиентов, и в первую очередь направить ресурсы на работу с ними. Для этого «Мегафон» провёл опрос своих клиентов, предложив им оценить уровень удовлетворённости качеством связи. По каждому клиенту, прошедшему опрос, были собраны технические показатели. 

Подготовьте исследование для компании «Мегафон» и проанализируйте, как зависит (и зависит ли) оценка, которую ставит клиент в опросе, от технических показателей, которые были собраны. 



## 1.1 Описание опроса

В ходе опроса компания «Мегафон» предложила своим клиентам оценить уровень удовлетворённости качеством связи по десятибалльной шкале (где 10 — это «отлично», а 1 — «ужасно»). Если клиент оценивал качество связи на 9 или 10 баллов, опрос заканчивался. Если клиент ставил оценку ниже 9, задавался второй вопрос — о причинах неудовлетворённости качеством связи с предоставленными пронумерованными вариантами ответа. Ответ можно было дать в свободном формате или перечислить номера ответов через запятую. Ниже вы можете ознакомиться с инфографикой по структуре опроса.

**Значения ответов пользователей**: 

- 1 - Недозвон, обрывы при звонках
- 2 - Время ожидания гудков при звонках
- 3 - Плохое качество связи в зданиях, торговых центраз
- 4 - Медленный мобильный интернет
- 5 - Медленная загрузка видео
- 6 - Затрудняюсь ответить
- 7 - Свой ответ

## 1.2 Codebook

`megafon.csv` содержит следующие значения: <br><br>
&nbsp;&nbsp;&nbsp;&nbsp; `user_id` — идентификатор абонента;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q1` — ответ на первый вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q2` — ответ на второй вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Total Traffic(MB)` — объем трафика передачи данных <sup>1 </sup>; <br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink Throughput(Kbps)` — средняя скорость «к абоненту» <sup>2 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Uplink Throughput(Kbps)`— средняя скорость «от абонента» <sup>3 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink TCP Retransmission Rate(%)` — частота переотправок пакетов «к абоненту» <sup>4 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Video Streaming Download Throughput(Kbps)` — скорость загрузки потокового видео <sup>5 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Video Streaming xKB Start Delay(ms)` — задержка старта воспроизведения видео <sup>6 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Web Page Download Throughput(Kbps)` — скорость загрузки web-страниц через браузер <sup>7 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Web Average TCP RTT(ms)` — пинг при просмотре web-страниц<sup>8 </sup>.<br>


<sup>1 </sup> — Насколько активно абонент использует мобильный интернет.<br>
<sup>2 </sup> — Считается по всему трафику передачи данных.<br>
<sup>3 </sup> — Считается по всему трафику передачи данных.<br>
<sup>4 </sup> — Чем выше, тем хуже. Если в канале возникает ошибка, пакет переотправляется. Снижается полезная скорость.<br>
<sup>5 </sup> — Чем выше, тем лучше — меньше прерываний и лучше качество картинки.<br>
<sup>6 </sup> — Сколько времени пройдёт между нажатием на кнопку Play и началом воспроизведения видео. Чем меньше это время, тем быстрее начинается воспроизведение.<br>
<sup>7 </sup> — Чем выше, тем лучше.<br>
<sup>8 </sup> — Чем меньше, тем лучше — быстрее загружаются web-страницы.<br>

Первый технический показатель представлен как сумма за период в одну неделю перед участием в опросе. Остальные технические показатели отображают среднее значение по данному признаку за период в одну неделю перед участием в опросе.

Цель исследования состоит в том, чтобы определить влияют ли технические показатели на удовлетворенность клиента сотовой связью. Основная гипотеза, которую мы выдвигаем заключается в том, что клиенты, высокооценивающие сотовую связь, имеют хорошие технические характеристики связи. И, наоборот, клиенты, у которых наблюдаются проблемы с мобильной связью, дают низкие оценки в опросе.

Так как эти аномальные данные не единичный случай, для того, чтобы исключить наименьшее количество данных, используем оптимизацию по процентилю.

```python

# Для удобства создадим список с названиями столбцов

col_name = []
for i in range(0, 11):
    col_name.append(survey.columns[i])
    
# Удалим из датафрейма аномальные значения

for col in col_name[3:]:
    survey = survey[survey[col] < survey[col].quantile(0.99)]

# Выведем описание данных 

df1 = survey.describe()
df1.to_csv('second_var.csv')

survey.describe()
```

Значения в данных все еще разнообразны, но без экстремальных аномалий. При этом заметим, что распредление оценок не изменилось. Половина пользователей ставят оценку не выше 7, а средняя оценка - 6. Так же было и в описании первоначального датафрейма. Это говорит, что наша оптимизация не исказила данных в целом.

```python
# Сгруппируем данные по первому вопросу

answer_Q1 = survey.groupby(['Q1'], as_index=False).agg({col_name[0]: ['count'],
                                                col_name[3]: ['mean'],
                                                col_name[4]: ['mean'],
                                                col_name[5]: ['mean'],
                                                col_name[6]: ['mean'],
                                                col_name[7]: ['mean'],
                                                col_name[8]: ['mean'],
                                                col_name[9]: ['mean'],
                                                col_name[10]: ['mean']
                                               }).round(2)
answer_Q1.columns = [col[0] for col in answer_Q1.columns.values]
answer_Q1
```

Визуализируем распределение частоты оценок пользователей на первый вопрос

```python
# Построим график распределения ответо на первый вопрос с помощью plotly express
pie_Q1 = px.pie(answer_Q1, names = 'Q1', values = 'user_id',
                title = 'Распределение ответов пользователей на первый вопрос', width = 1000, height = 500)
pie_Q1.show()
pie_Q1.write_image("pie_Q1.png") 
```

Самыми частыми оценками оказываются оценка 10 и 1. Посторим графики по каждому техническому показателю и проверим, есть ли зависимость оценок от качества связи у абонента. 

```python
# Создаем интерактивный график 

scatter_Q1 = go.Figure()
start_feature = 'Total Traffic(MB)'
for col in col_name[3:]:
    visible = False
    if (col == start_feature):
        visible = True
        title_text = {'title' : start_feature}
        scatter_Q1 = go.Figure(layout=title_text)
    scatter_Q1.add_trace(
        go.Scatter(x=answer_Q1['Q1'], y=answer_Q1[col], visible=visible, mode='markers+lines', name=col)
    )
```

```python
# Создаем список словарей для информативного отображения блока слайдера

mysteps = []
for i in range(len(col_name[3:])):
    tmp = []
    for j in range(len(col_name)):
        if (i == j):
            tmp.append(True)
        else:
            tmp.append(False)
    step = dict(
        method = 'update',
        args = [
            {'visible': tmp},
            {'title.text' : scatter_Q1.data[i].name},
        ],
        label = scatter_Q1.data[i].name
    )
    mysteps.append(step)
```

```python
# Добавляем данные для слайдера на график, прописываем подписи и заголовок

scatter_Q1.update_layout(
    sliders=[dict(
        steps=mysteps
    )]    
)
scatter_Q1.update_layout({'xaxis': {'dtick': [1,11]},
                          'title' : 'Средние значения технических показателей в зависимости от оценки пользователя',
                          'xaxis_title' : 'score',
                          'yaxis_title' : 'average value',
                          'height' : 400,
                          'width' : 1000
                         })

scatter_Q1.show() 
```

На графиках некоторых показателей четко прослеживается закономерность. Например, у пользователей поставивших оценку 6 и выше в среднем значительно лучше показатель задержки старта потокового видео (Video Streaming xKB Start Delay(ms)). Проверим наши наблюдения, посчитаем, есть ли корреляция между тем, как пользователь оценивает связь и какие у него технические характеристики связи.

```python
# Посчитаем корреляцию относительно ответов на первый вопрос для всех технических характеристик

for col in col_name[3:]:
    print(col, ': ', round(answer_Q1['Q1'].corr(answer_Q1[col]), 2))
```

Наблюдается большая корреляция между высотой оценки и техническими показателями у пользователя. Все показатели, кроме объема траффика коррелируют с оценкой. Наибольшая корреляция выявлена у показателя Video Streaming xKB Start Delay(ms) и Web Average TCP RTT(ms). Это означает, что эти два показателя наибольшим образом влияют на то, как пользователь оценивает качество интернет соединения в целом. Действительно, скорость начала воспроизведения видео и скорость ответа на действие пользователя это то, с чем пользователь стакливается чаще всего и может максимально заметно влиять на восприятие пользователя о том, насколько быстро "летает интернет". 

Но не стоит забывать, что это корреляция между высотой оценки и средним техническим показателем в группе позльзователей, поставивших ту или иную оценку. Поэтому результаты очень приблизительные, хоть и смотрятся логичными.

Проверим полученные данные. Проведем статистический тест для показателя, который имеет максимально высокую корреляцию с оценкой на первый вопрос: Video Streaming xKB Start Delay(ms). Действительно ли это характеристика влияет на общую удовлетворенность клиентов связью.

### 2.2 Статистический тест

**Показатель Video Streaming xKB Start Delay и оценки пользователей на первый вопрос.**

Выдвенем две гипотезы:

- Н0 = Скорость загрузки видео не влияет на удовлетворенность клиентов связью,
- Н1 = Скорость загрузки видео положительно влияет на удовлетворенность клиентов связью.

Для начала построим график распределения показателя Video Streaming xKB Start Delay(ms) для двух групп пользователей: поставивших оценку 9 и 10, назовём их "Довольные пользователи", и для пользователей, поставивших оценку 8 и ниже, назовём их "Недовольные пользователи".

```python
# График распределения показателя Video Streaming xKB Start Delay(ms) для пользователей в двух группах

sns.distplot(survey[survey['Q1'] > 8]['Video Streaming xKB Start Delay(ms)'], color='red', label='Довольные пользователи')
sns.distplot(survey[survey['Q1'] <= 8]['Video Streaming xKB Start Delay(ms)'], color='blue', label='Недовольные пользователи')
plt.legend()
plt.title('Гистограммы распределения Video Streaming xKB Start Delay(ms) \n для двух групп абонентов')
plt.xlabel('Значения метрики')
plt.ylabel('Частота')
plt.show()
```

Оба графика скошены влево, визуально мы можем определить, что у большего числа пользователей в группе "Довольные пользователи" показатели характеристики Video Streaming xKB Start Delay(ms) меньше, чем во второй. (В данном случае, чем меньше показатель, тем лучше качество интернет соединения). Прооверим эти подозрения, построив доверительные интервалы двух групп.

```python
# Доверительный интервал в группе пользователей, высоко оценивших мобильную связь

stm.DescrStatsW(survey[survey['Q1'] > 8]['Video Streaming xKB Start Delay(ms)']).tconfint_mean()
```

```python
# Доверительный интервал в группе пользователей, низко оценивших мобильную связь

stm.DescrStatsW(survey[survey['Q1'] <= 8]['Video Streaming xKB Start Delay(ms)']).tconfint_mean()
```

Доверительный интервалы не пересекаются. Это означает, что мы можем отвергнуть нулевую гипотезу в пользу альтернативной гипотезы. Пользователи, оценивающие сотовую связь положительно, имеют лучшие технические показатели по загрузке потокового видео.

Для быстрого анализа данных создадим функции, которые будем использовать для решения этих задач.

```python
# Функция построения гистограммы сравниваемых групп

def myhist(df_A, df_B,  feature):

    # Отобразим на гистограмме распределение значений интересующей нас характеристики у двух групп пользователей

    sns.distplot(df_A[feature], color='red', label='Недовольные пользователи')
    sns.distplot(df_B[feature], color='blue', label='Довольные пользователи')
    plt.legend()
    plt.title('Гистограммы распределения {} \n для двух групп абонентов'.format(feature))
    plt.xlabel('Значения метрики')
    plt.ylabel('Частота')
    
    return plt.show()   
```

```python
# Функция бутстреп

def bootstrap(dat, n=1000, N_TRIAL = 1000):
    
    new_dat = []
    for _ in range (N_TRIAL):
        subsample = np.random.choice(dat, size = (n,))
        stat = np.mean(subsample)
        new_dat.append(stat)

    return new_dat
```

```python
# Функция построения гистогаммы распределения с использованием бутстреп

def hist_boot(group_A, group_B):
    
    fig, ax = plt.subplots()
    ax.hist(group_A, bins = 20, alpha = 0.3)
    ax.hist(group_B, bins = 20, alpha = 0.3)
    ax.legend(["Недовольные клиенты", "Довольные клиенты"])

    ax.axvline(np.percentile(group_A, 2.5), linestyle = "-", color="r")
    ax.axvline(np.percentile(group_A, 97.5), linestyle = "-", color="g")

    ax.axvline(np.percentile(group_B, 2.5), linestyle = "--", color="r")
    ax.axvline(np.percentile(group_B, 97.5), linestyle = "--", color="g")
```

```python
# Функция разницы средних

def dif_mean(df_A, df_B, feature):
    dif_data = np.mean(df_B[feature]) - np.mean(df_A[feature])
    dif = []
    for i in range(1000):
      sample_A = df_A[feature].sample(3000, replace=True)
      sample_B = df_B[feature].sample(3000, replace=True)
      dif.append(np.mean(sample_B) - np.mean(sample_A))
        
    return [dif, dif_data]
```

```python
# Гистограмма разницы средних

def hist_dif(dif, dif_data, feature):
    
    p = (dif - np.mean(dif) >= dif_data).sum()/len(dif)
    
    fig = go.Figure(data=[
        go.Histogram(x=dif, nbinsx=100, marker_color= 'lightblue'),
    ])
    fig.add_vline(x=np.percentile(dif, 2.5), line_color = 'red', line_width=1)
    fig.add_vline(x=0, line_color = 'blue', line_dash = 'dot', line_width=1)
    fig.add_vline(x=np.percentile(dif, 97.5), line_color = 'green', line_width=1)
    fig.update_layout(title_text ='Распределение средних для {}'.format(feature))
    fig.show()
    print("min =", np.percentile(dif, 2.5), ', max =',np.percentile(dif, 97.5), 
      '\n p-value =', p)
```

#### 3.2.1 Показатель Downlink Throughput(Kbps)

**Группа пользователей, отметивших медленный мобилный интернет:**

- Н1: Средняя скорость "к абоненту" ниже у тех пользователей, что отметили эту проблему при оценке Pa < Pb

Построим гистограмму распределения показателей Downlink Throughput(Kbps) для группы пользователей, которые указали в своих ответах причину "Медленный мобильный интернет", и для группы пользователей, которые эту причину не указывали.


```python
# Разделим пользователей на группы довольных и недовольных по ответу: "Медленный интернет"

df_A = filter_df('4')[0] # Недовольные пользователи
df_B = filter_df('4')[1] # Довольные пользователи

# Построим гистограмму распределения показателей Downlink Throughput(Kbps) у двух групп пользователей,
# указавших эту проблему в ответах и не указавших

myhist(df_A, df_B, 'Downlink Throughput(Kbps)')
```

На гистограмме заметно сдвинуто влево распределение значений метрики у пользователей в обеих группах. Данных в нашей выборке недостаточно. Проверим эту гипотезу с помощью метода бутсреп.

```python
group_A = bootstrap(df_A['Downlink Throughput(Kbps)'], n = 1000, N_TRIAL = 3000) # Недовольные пользователи
group_B = bootstrap(df_B['Downlink Throughput(Kbps)'], n = 1000, N_TRIAL = 3000) # Довольные пользователи
```

Построим графики распределения в полученных с помощью бутстреп данных

```python
hist_boot(group_A, group_B)
```

Доверительные интервалы не пересекаются. Для того, чтобы убедиться наверняка, в том, что различия между полученными данными есть визуализируем разницу средних и вычислим p-value.

```python
dif = dif_mean(df_A, df_B, 'Downlink Throughput(Kbps)')[0] # список разности средних сгенирированных случайных данных
dif_data = dif_mean(df_A, df_B, 'Downlink Throughput(Kbps)')[1] # разность средних в исходных данных
```

```python
# Построим график распределения средних

hist_dif(dif, dif_data, 'Downlink Throughput(Kbps)')
```

Проверим значение p-value с помощью теста Манна-Уитни:

```python
p_value = stats.mannwhitneyu(df_A['Downlink Throughput(Kbps)'], df_B['Downlink Throughput(Kbps)'],
                             alternative='two-sided')[1]
print('p-value = ', p_value)
```

Распределение разницы средних не включает ноль, а это означает, что разница между группами пользователей есть. Так же значение p-value меньше уровня значимости 0.05.

```python
# Разделим пользователей на группы довольных и недовольных по ответу: "Медленная загрузка видео"

df_A = filter_df('5')[0] 
df_B = filter_df('5')[1]

# Построим гистограмму распределения показателей Video Streaming xKB Start Delay(ms) у двух групп пользователей,
# указавших эту проблему в ответах и не указавших

myhist(df_A, df_B, 'Video Streaming Download Throughput(Kbps)')
```

На гистограмме заметно сдвинуто влево распределение значений метрики у пользователей в обеих группах. Данных в нашей выборке недостаточно. Проверим эту гипотезу с помощью метода бутсреп.

```python
group_A = bootstrap(df_A['Video Streaming Download Throughput(Kbps)'], n = 1000, N_TRIAL = 3000)
group_B = bootstrap(df_B['Video Streaming Download Throughput(Kbps)'], n = 1000, N_TRIAL = 3000)
```


Построим графики распределения в полученных с помощью бутстреп данных.

```python
hist_boot(group_A, group_B)
```

Доверительные интервалы не пересекаются. Для того, чтобы убедиться наверняка, в том, что различия между полученными данными есть визуализируем разницу средних и вычислим p-value.

```python
dif = dif_mean(df_A, df_B, 'Video Streaming Download Throughput(Kbps)')[0]
dif_data = dif_mean(df_A, df_B, 'Video Streaming Download Throughput(Kbps)')[1]
```

```python
hist_dif(dif, dif_data, 'Video Streaming Download Throughput(Kbps)')
```

Проверим значение p-value с помощью теста Манна-Уитни:

```python
p_value = stats.mannwhitneyu(df_A['Video Streaming Download Throughput(Kbps)'],
                             df_B['Video Streaming Download Throughput(Kbps)'],
                             alternative='two-sided')[1]
print('p-value = ', p_value)
```

Распределение разницы средних не включает ноль, а это означает, что можно утверждать, что разница между группами пользователей есть. Так же значение p-value меньше уровня значимости 0.05.

Наши вычисления показывают, что нулевую гипотезу стоит отбросить. И скорость загрузки потокового видео влияет на удовлетворенность пользователей скоростью загрузки видео.
